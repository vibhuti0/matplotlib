# Multiple Points

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# Default X-points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4, 8, 6, 5, 2, 6])

plt.plot(ypoints)
plt.show()
```

# Markers
Used to specify each point in a graph.We can use different types of markers to specify the points for eg. point,diamond,square,star etc.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'D')
plt.show()
```

ms or marker size is used to specify the size of the marker.
mfc is marker face color
mec is marker edge color
we can also give hexadecimal value for color.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--',ms = 20,mfc = 'r',mec = 'b')
plt.show()
```

The 5 different types of plot in pyplot module of matplotlib are:

**Line** 

**Bar**

**Scatter**

**Histogram**

**Pie chart**

# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted',c='r',linewidth='5.5')
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

# Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Honda", "Hyundai", "BMW", "Tesla"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y,color ="pink")
plt.show()
```

# Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([45, 65, 25, 35])

plt.pie(y)
plt.show() 
```
